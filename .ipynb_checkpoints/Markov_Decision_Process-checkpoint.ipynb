```python
def create_state(length):
#     print(length)
    cvp_value = length['CVP']
    urine_value = length['Urine']
    age_value = length['CVP']
#     numbers = cvp_values.values
        
    if ((8 < cvp_value < 13) and (((age_value <= 12) and (urine_value > 2)) or ((age_value > 12) and (urine_value > 1)))):
        return 0 #improve
#     elif ((cvp_value == 8) and (((age_value <= 12) and (urine_value == 2)) or ((age_value > 12) and (urine_value == 1)))):
#         return 1
    elif (((cvp_value < 8) or (cvp_value > 13)) and (((age_value <= 12) and (urine_value < 2)) or ((age_value > 12) and (urine_value < 1)))):
        return 1 #detroit
    else:
        return 2 #stable
```

```python
df['state'] = df.apply(create_state, axis=1)
df.head(20)
```

```python
def transition_matrix(transitions):
    n = 1+ max(transitions) #number of states

    M = [[0]*n for _ in range(n)]

    for (i,j) in zip(transitions,transitions[1:]):
        M[i][j] += 1

    #now convert to probabilities:
    for row in M:
        s = sum(row)
        if s > 0:
            row[:] = [f/s for f in row]
    return M
```

```python
transitionMatrix = np.array(transition_matrix(df['state'])).round(4)
```

```python
# Probabilities matrix (transition matrix)
print("Transition Matrix:")
transitionMatrix
```
