```python
def create_state(value):
    cvp_value = value['CVP']
    urine_value = value['Urine']
    age_value = value['Age']
    bp_value = value['ABP (S)']
    
    if ((8 < cvp_value < 13) and (80 <= bp_value <= 120) and (((age_value <= 12) and (urine_value > 2)) or ((age_value > 12) and (urine_value > 1)))):
        return 'Imporovement'
    elif (((cvp_value < 8) or (cvp_value > 13)) and (bp_value < 70) and (((age_value <= 12) and (urine_value < 2)) or ((age_value > 12) and (urine_value < 1)))):
        return 'Detroite'
    else:
        return 'Stable'
```

```python
def transition_matrix(transitions):
    
#     int_map = map(, transitions)
#     int_list = list(int_map)
    n = 1+ max(transitions) #number of states

    M = [[0]*n for _ in range(n)]

    for (i,j) in zip(transitions,transitions[1:]):
        M[i][j] += 1

    #now convert to probabilities:
    for row in M:
        s = sum(row)
        if s > 0:
            row[:] = [f/s for f in row]
    return M
```

```python
# transitionMatrix = transition_matrix(tmp_df['state'])
transitionMatrix = np.array(transition_matrix(tmp_df['state'])).round(4)
```
